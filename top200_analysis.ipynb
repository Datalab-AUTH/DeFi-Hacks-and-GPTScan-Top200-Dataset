```python
output_file = "/Users/elenimandana/Projects/Thesis/MSc-thesis/data/contracts/top200/top200_sampled_functions.json"

with open(output_file, 'r') as file:
    data = json.load(file)
```

```python
# Count entries
count = len(data)
print(f"Number of entries: {count}")
```

```python
# Calculate code lengths
code_lengths = [len(entry["code"]) for entry in data]
if code_lengths:
    print(f"Count: {len(code_lengths)}")
    print(f"Total length: {sum(code_lengths)}")
    print(f"Average length: {sum(code_lengths) / len(code_lengths):.2f}")
    print(f"Minimum length: {min(code_lengths)}")
    print(f"Maximum length: {max(code_lengths)}")

    plt.figure(figsize=(10, 6))
    plt.hist(code_lengths, bins=30, edgecolor='black')
    plt.title('Function Length Distribution')
    plt.xlabel('Length of Code (characters)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Word frequency analysis
all_code = " ".join(entry["code"] for entry in data)
words = all_code.split()
word_freq = Counter(words)

print("Top 10 most common words:")
for word, count in word_freq.most_common(10):
    print(f"{word}: {count}")
```

```python
# Word cloud
txt = " ".join(words)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(txt)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
